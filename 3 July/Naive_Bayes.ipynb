```python
#Session Task - (3-7-24)

#Use iris flower dataset from sklearn library and use Naive bayes models to measure the performance of each.
#1. Accuracy_score
#2. Classification_report
#3. Confusion_matrix



import pandas as pd
import numpy as np
```

```python
# Iris Dataset
```

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
label_encoder
```

```python
df['species'] = label_encoder.fit_transform(df['species'])
```

```python
df
```

```python
X= df.drop(['species'], axis = 1)
```

```python
X
```

```python
y= df['species']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test , y_train ,y_test = train_test_split(X,y, test_size=0.3, random_state = 45 )
```

```python
X_train
```

```python
y_test
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train , y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
gnb.score (X_test ,y_test)
```

```python
pr = gnb.predict([[4.9,3.1,1.5,0.1]])
pr
label_encoder.inverse_transform(pr)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
acc= accuracy_score(y_test , y_pred)
acc
```

```python
cl = classification_report(y_test , y_test)
```

```python
cl
```

```python
cm=confusion_matrix(y_test,y_pred)
```
