```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(solver='saga', tol=1e-2, max_iter=200)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

param_distributions = {
    'C': uniform(0.001, 100),  # Example range for regularization strength
    'penalty': ['l1', 'l2']  # Regularization type
}

random_search = RandomizedSearchCV(estimator=logistic, param_distributions=param_distributions, n_iter=10, cv=5)
random_search.fit(X_train, y_train)

best_params = random_search.best_params_
best_score = random_search.best_score_
print("Best hyperparameters:", best_params)
print("Best accuracy score:", best_score * 100)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}

grid_search = GridSearchCV(estimator=logistic, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_params_grid = grid_search.best_params_
best_score_grid = grid_search.best_score_
print("Best hyperparameters (GridSearchCV):", best_params_grid)
print("Best accuracy score (GridSearchCV):", best_score_grid * 100)
```
